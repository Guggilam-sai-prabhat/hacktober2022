```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
input_image = Image.open("trio.jpeg")
```

```python
pixel_map = input_image.load()
width,height = input_image.size
for i in range(width):
    for j in range(height):
        r,g,b = input_image.getpixel((i,j))
        gray_scale = (0.299*r + 0.587*g + 0.114*b)
        #gray_scale2 = (r+g+b)/3
        pixel_map[i,j] = (int(gray_scale),int(gray_scale),int(gray_scale))
input_image.save("grayscale.png",format = "png")
```

```python
plt.imshow(input_image)
```

```python
import numpy as np
input_image = Image.open("trio.jpeg")
plt.imshow(input_image)
input_image = np.asarray(input_image)
print(input_image.shape)
```

```python
row,col,plane = input_image.shape
x,y = 2,2
b_plane = input_image[:,:,0]
g_plane = input_image[:,:,1]
r_plane = input_image[:,:,2]

reb_plane = b_plane[1::x,1::x]
reg_plane = g_plane[1::x,1::x]
rer_plane = r_plane[1::x,1::x]

resize_img = np.zeros((row//x,col//y,plane),np.uint8)

resize_img[:,:,0] = reb_plane
resize_img[:,:,1] = reg_plane
resize_img[:,:,2] = rer_plane

plt.imshow(resize_img)
print(resize_img.shape)
```
